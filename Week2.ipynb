```python
# Jawab : Tampilkan image dalam channel Red-Green dan Green-Blue saja!
imgRoot = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/image.jpeg')

# Green-Blue
getImg1 = cv.cvtColor(imgRoot,cv.COLOR_BGR2RGB)
getImg1[:,:,0]=0
plt.imshow(getImg1)
```

```python
# Red-Green
getImg2 = cv.cvtColor(imgRoot,cv.COLOR_BGR2RGB)
getImg2[:,:,2]=0
plt.imshow(getImg2)
```

```python
# Jawab : Tampilkan image baris ke 10-100, kolom 10-100!
getImg3 = imgRoot
plt.imshow(getImg3[10:100,10:100])
```
