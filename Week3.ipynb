```python
# 1. Implementasikan transformasi contrast pada Google Colaboratory menggunakan formula yang terdapat pada bagian Ulasan Teori,
print('Mengubah kontras dan tingkat kecerahan citra')
print('================================')
try:
  brightness = float(input('Input nilai kecerahan: '))
  contrast = float(input('Input nilai kontras: '))
except ValueError:
  print('Error, not a number')

imgTask1 = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/female.tiff')

brightness_image = np.zeros(imgTask1.shape, imgTask1.dtype)

for y in range(imgTask1.shape[0]):
  for x in range(imgTask1.shape[1]):
    for c in range(imgTask1.shape[2]):
      brightness_image[y,x,c] = np.clip(imgTask1[y,x,c] + brightness * contrast, 0, 255) #rumus

final_frame = cv.hconcat((imgTask1,brightness_image))
cv2_imshow(final_frame)
```

```python
# 2. Implementasikan transformasi logarithmic brightness pada Google Colaboratory menggunakan formula yang terdapat pada bagian Ulasan Teori
print('Mengubah tingkat kecerahan citra dengan transformasi log')
print('================================')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

imgTask2 = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/female.tiff')
brightness_image = np.zeros(imgTask2.shape, imgTask2.dtype)

for y in range(imgTask2.shape[0]):
  for x in range(imgTask2.shape[1]):
    for c in range(imgTask2.shape[2]):
      brightness_image[y,x,c] = np.clip(brightness*(np.log(1+imgTask2[y,x,c])),0,255)#rumus

final_frame = cv.hconcat((imgTask2,brightness_image))
cv2_imshow(final_frame)
```

```python
# 3. Implementasikan transformasi grayscale menggunakan metode averaging, lightness, dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada bagian Ulasan Teori,

# averaging
print("AVERAGING")

imgLeftTask31 = cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff'),cv.COLOR_BGR2RGB)
imageRightTask31 = cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff'),cv.COLOR_BGR2RGB) 

imageRightTask31[:] = np.sum(imageRightTask31, axis=-1, keepdims=1)/3

fig, axarr = plt.subplots(1,2)
axarr[0].imshow(imgLeftTask31)
axarr[1].imshow(imageRightTask31)
```

```python
# lightness
print("LIGHTNESS")

imgLeftTask32 = cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff'),cv.COLOR_BGR2RGB)
imageRightTask32 =  cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff'), cv.COLOR_BGR2RGB)

imageRightTask32[:] = (np.max(imageRightTask32, axis=-1, keepdims=1))/2 + (np.min(imageRightTask32, axis=-1, keepdims=1))/2

fig, axarr = plt.subplots(1,2)
axarr[0].imshow(imgLeftTask31)
axarr[1].imshow(imageRightTask32)
```

```python
# Luminance
print("LUMINANCE")

imgLeftTask32 = cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff'),cv.COLOR_BGR2RGB)
imageRightTask32 =  cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff'), cv.COLOR_BGR2RGB)

resultLuminance = (0.21 * imageRightTask32[:,:,0]) + (0.72 * imageRightTask32[:,:,1]) + (0.07 * imageRightTask32[:,:,2])

fig, axx = plt.subplots(1,2)
axx[0].imshow(imgLeftTask31)
axx[1].imshow(resultLuminance,cmap='gray')
```

```python
# 4. Tampilkanlah warna tertentu pada citra, dan ubah warna lain menjadi grayscale. Misal, tampilkan warna merah pada citra masukan dan ubah bagian lain yang tidak berwarna merah menjadi grayscale

imgTask4 = cv.cvtColor(cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff'),cv.COLOR_BGR2RGB)

lowRange = np.array([0,0,0])
upRange = np.array([255,110,120])
maskImg = cv.inRange(imgTask4,lowRange,upRange)
result = cv.bitwise_and(imgTask4,imgTask4,mask=maskImg)

imgTask5 = cv.cvtColor(imgTask4,cv.COLOR_BGR2GRAY)
imgBgr = cv.cvtColor(imgTask5,cv.COLOR_GRAY2BGR)
results = cv.bitwise_or(imgBgr,result)

fig, axx = plt.subplots(1,3)
axx[0].imshow(imgTask4)
axx[1].imshow(result)
axx[2].imshow(results)
```
